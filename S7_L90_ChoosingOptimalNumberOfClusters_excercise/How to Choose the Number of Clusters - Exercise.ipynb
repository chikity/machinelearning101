## Selecting the number of clusters

### WCSS

Use the ingerated <i>sklearn</i> method <i> 'inertia_' </i>.

```python
clusters = range(1,21)
wcss = []

for i in clusters:
    model = KMeans(i)
    model.fit(x)
    wcss.append(model.inertia_)
```

Write a loop that calculates and saves the WCSS for any number of clusters from 1 up to 10 (or more if you wish).

```python
plt.plot(clusters, wcss)
plt.title('The Elbow Method', size = 20)
plt.xlabel('Clusters')
plt.ylabel('WCSS')
plt.xticks(clusters)
```

```python
def plot_clusters(data: pd.DataFrame, number_of_clusters: int):
    kmeans = KMeans(number_of_clusters)
    kmeans.fit(x)
    
    results = kmeans.predict(x)
    df = data.copy()
    df['Cluster'] = results

    plt.scatter(df['Longitude'], df['Latitude'],
                c = df['Cluster'],
                cmap= 'rainbow')
```

### The Elbow Method

```python
plot_clusters(data, 4)
```
